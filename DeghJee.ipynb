```python
import pandas as pd
import math
import random
df=pd.DataFrame()
```

```python
for i in range(100,1001,5):
    data={
    'event':['shadi'],
    'total_people':[i],
    'city':['karachi'],
    'rice':[1],
    'curry':[0],
    'sea_food':[0],
    'bbq':[0],
    'dessert':[0],
    'dish_name':['chinese'],
    'dish_type':['rice'],
    'amount':[math.ceil((i*(random.randrange(10,12)))/100)]
    }
    dict_data= pd.DataFrame.from_dict(data)
    df=df.append(dict_data,sort=False)
    df.to_csv('deghjee.csv')
df
```
